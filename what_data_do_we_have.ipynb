```python
import pandas as pd


# Load in legacy and updated subscription data

base_path = '/Users/metinyarici/Library/CloudStorage/OneDrive-SharedLibraries-LincolnCityFC/Player Recruitment - Data Science/statsbomb_things/'
df = pd.read_csv(base_path+'data/player_season_stats.csv')


# '''
# Pre-processing
# '''

# # Clean col names in df
# df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()

# # Remove 'player_season' from any column names that include it
# df.rename(columns=lambda x: x.replace('player_season_', ''), inplace=True)

# # Strip and lowercase and add underscores where spaces exists between words for values in 'primary_position'
# df['primary_position'] = df['primary_position'].str.strip().str.lower().str.replace(' ', '_')

# # Strip and lowercase values in competition_name, also replace spaces with '_'
# df['competition_name'] = df['competition_name'].str.strip().str.lower().str.replace(' ', '_')

# # Strip and lowercase values in player_name
# df['player_name'] = df['player_name'].str.strip().str.lower().str.replace(' ', '_')

# # Create goals - Xg fetaure
# df['np_goals_less_xg_90'] = df['npga_90']-df['assists_90']-df['np_xg_90']


# # Keep only rows with highest value in 'minutes' if multiple rows share the same player_id, competition_id and season_id

# # Assuming your DataFrame is named 'df'
# df = df.sort_values(['player_id', 'competition_id', 'season_id', 'minutes'], ascending=[True, True, True, False])

# # Keep only the first occurrence of each group after sorting, which will have the highest 'minutes'
# df = df.drop_duplicates(subset=['player_id', 'competition_id', 'season_id'], keep='first')

# # Optionally, if you want to reset the index after dropping duplicates
# df = df.reset_index(drop=True)

'''
Search for particular player
'''


particular_id = 31663 #Baccay

# Find all data for this player
particular_player_df = df[df['player_id'] == particular_id]
particular_player = particular_player_df['player_name'].iloc[0]
# For each unique combination of season_name and competition_name, display table with minutes, appearances, and team_name
# Group by season_name and competition_name
grouped = particular_player_df.groupby(['season_name', 'competition_name'])

# Iterate over each group to create a separate table
for (season, competition), group in grouped:
    # Select only the required columns
    table = group[['minutes', 'appearances', 'team_name']]

    # Optionally, you can sort the table if needed, e.g., by minutes descending
    table = table.sort_values('minutes', ascending=False)

    # Display the table
    print(f"Season: {season}, Competition: {competition}")
    print(table)
    print("\n" + "="*50 + "\n")  # Just for visual separation between tables

# Convert grouped DataFrameGroupBy object to a regular DataFrame
table = grouped[['season_name', 'competition_name', 'minutes', 'appearances', 'team_name']]
if isinstance(table, pd.core.groupby.generic.DataFrameGroupBy):
    table = table.obj  # Get the underlying DataFrame from the GroupBy object

# Sort the table by season_name and then by minutes descending
table = table.sort_values(['season_name', 'minutes'], ascending=[True, False])

# Begin LaTeX table code
latex_code = "\\begin{table}[h]\n"
latex_code += "\\centering\n"
latex_code += "\\begin{tabular}{|l|l|l|l|l|}\n"
latex_code += "\\hline\n"
latex_code += "Season & Competition & Minutes & Appearances & Team Name \\\\\n"
latex_code += "\\hline\n"

# Convert each row to LaTeX format
for _, row in table.iterrows():
    latex_code += f"{row['season_name']} & {row['competition_name'].title().replace('_',' ')} & {row['minutes']} & {row['appearances']} & {row['team_name']} \\\\\n"

# End LaTeX table
latex_code += "\\hline\n"
latex_code += "\\end{tabular}\n"
latex_code += f"\\caption{{Available data for {particular_player.replace('_',' ').title()}}}\n"
latex_code += f"\\label{{tab:data_avaialable}}\n"
latex_code += "\\end{table}\n"

# Print the LaTeX code
print(latex_code)
```
